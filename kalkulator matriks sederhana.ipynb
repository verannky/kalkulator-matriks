```python
#RUMUS
def pilihmenu():
  pilih_menu = input('Pilih menu: ')
  if pilih_menu=='1':
    print('=== Perkalian skalar ===')
    print()
    print('Pilih matriks:')
    print('a. Matriks A')
    print('b. Matriks B')
    perkalianskalar()
    hitunglagi()
  elif pilih_menu=='2':
    print('=== Penjumlahan matriks ===')
    print()
    print('Pilih matriks:')
    print('a. Matriks A+B')
    print('b. Matriks B+A')
    penjumlahan()
    hitunglagi()
  elif pilih_menu=='3':
    print('=== Pengurangan matriks ===')
    print()
    print('Pilih matriks:')
    print('a. Matriks A-B')
    print('b. Matriks B-A')
    pengurangan()
    hitunglagi()
  elif pilih_menu=='4':
    print('=== Perkalian matriks ===')
    print()
    print('Pilih matriks:')
    print('a. Matriks AxB')
    print('b. Matriks BxA')
    perkalian()
    hitunglagi()
  elif pilih_menu=='5':
    print('=== Determinan matriks ===')
    print()
    print('Pilih jenis matriks:')
    print('a.Ordo matriks 2x2')
    print('b.Ordo matriks 3x3')
    print('c.kembali ke menu?')
    determinan()
    hitunglagi()
  elif pilih_menu=='6':
    print('=== Keluar ===')
    exit()
  else:
    print('Pilihan menu salah')
    print()
    menu()
    
#Rumus skalar
def rumusskalar(matrix):
    scalar = int(input("Masukkan nilai skalar: "))
    result = [[0 for i in range(len(matrix[0]))] for i in range(len(matrix))]
    for i in range(len(matrix)):
        for j in range(len(matrix[0])):
            result[i][j] = matrix[i][j] * scalar
    return result

def perkalianskalar():
    pilihan= input('Pilihan: ')
    if pilihan in ['a', 'A']:
        hitungskalarA()
    elif pilihan in ['b', 'B']:
        hitungskalarB()
    elif pilihan in ['c', 'C']:
        menu()
    else:
        print('Pilihan salah')
def hitungskalarA():
    result = rumusskalar(matriks_A)
    print('Hasil perkalian skalar matriks A :')
    for i in result:
        print(i)
def hitungskalarB():
    result = rumusskalar(matriks_B)
    print('Hasil perkalian skalar matriks B :')
    for i in result:
        print(i)
        
#Rumus penjumlahan
def penjumlahanmatriks(matrix1, matrix2):
    result = [[0 for i in range(len(matrix1[0]))] for i in range(len(matrix1))]
    for i in range(len(matrix1)):
        for j in range(len(matrix1[0])):
            result[i][j] = matrix1[i][j] + matrix2[i][j]
    return result

def penjumlahan():
    pilihan= input('Pilihan: ')
    if pilihan in ['a', 'A']:
        printtambahAB()
    elif pilihan in ['b', 'B']:
        printtambahBA()
    else:
        print('Pilihan salah')
        
def hitungjumlahAB(matrix1, matrix2):
    if len(matrix1) != len(matrix2) or len(matrix1[0]) != len(matrix2[0]):
        print('Error: Dimensi matriks berbeda.')
        return
    jumlahAB = penjumlahanmatriks(matrix1, matrix2)
    print('Penjumlahan matriks A + matriks B :')
    for i in jumlahAB:
            print(i)
            
def hitungjumlahBA(matrix1, matrix2):
    if len(matrix1) != len(matrix2) or len(matrix1[0]) != len(matrix2[0]):
        print('Error: Dimensi matriks berbeda.')
        return
    jumlahAB = penjumlahanmatriks(matrix1, matrix2)
    print('Penjumlahan matriks B + matriks A :')
    for i in jumlahAB:
            print(i)
              
def printtambahAB():
    hitungjumlahAB(matriks_A,matriks_B)
              
def printtambahBA():
    hitungjumlahBA(matriks_B,matriks_A)
    
#Rumus pengurangan
def penguranganmatriks(matrix1, matrix2):
    result = [[0 for i in range(len(matrix1[0]))] for i in range(len(matrix1))]
    for i in range(len(matrix1)):
        for j in range(len(matrix1[0])):
            result[i][j] = matrix1[i][j] - matrix2[i][j]
    return result

def pengurangan():
    pilihan= input('Pilihan: ')
    if pilihan in ['a', 'A']:
        printkurangAB()
    elif pilihan in ['b', 'B']:
        printkurangBA()
    else:
        print('Pilihan salah')
              
def hitungkurangAB(matrix1, matrix2):
    if len(matrix1) != len(matrix2) or len(matrix1[0]) != len(matrix2[0]):
        print('Error: Dimensi matriks berbeda.')
        return
    penguranganAB= penguranganmatriks(matrix1, matrix2)
    print('Pengurangan matriks A - matriks B :')
    for i in penguranganAB:
        print(i)
              
def hitungkurangBA(matrix1, matrix2):
    if len(matrix1) != len(matrix2) or len(matrix1[0]) != len(matrix2[0]):
        print('Error: Dimensi matriks berbeda.')
        return
    penguranganBA= penguranganmatriks(matrix1, matrix2)
    print('Pengurangan matriks B - matriks A :')
    for i in penguranganBA:
        print(i)

def printkurangAB():
    hitungkurangAB(matriks_A,matriks_B)
              
def printkurangBA():
    hitungkurangBA(matriks_B,matriks_A)
              
#Rumus perkalian
def perkalianmatriks(matrix1, matrix2):
    result = [[0 for i in range(len(matrix1[0]))] for i in range(len(matrix1))]
    for i in range(len(matrix1)):
        for j in range(len(matrix2[0])):
            for k in range(len(matrix2)):
                result[i][j] += matrix1[i][k] * matrix2[k][j]
    return result

def perkalian():
    pilihan= input('Pilihan: ')
    if pilihan in ['a', 'A']:
        printkaliAB()
    elif pilihan in ['b', 'B']:
        printkaliBA()
    else:
        print('Pilihan salah')
        
def hitungkaliAB(matrix1,matrix2):
    if len(matrix1[0]) != len(matrix2):
        print("Error: Matrix tidak dapat dikalikan.")
        return
    perkalianAB = perkalianmatriks(matrix1,matrix2)
    print('Perkalian matriks A x matriks B :')
    for i in perkalianAB:
        print(i)
        
def hitungkaliBA(matrix1,matrix2):
    if len(matrix1[0]) != len(matrix2):
        print("Error: Matrix tidak dapat dikalikan.")
        return
    perkalianBA = perkalianmatriks(matrix1,matrix2)
    print('Perkalian matriks B x matriks A :')
    for i in perkalianBA:
        print(i)

def printkaliAB():
    hitungkaliAB(matriks_A,matriks_B)
              
def printkaliBA():
    hitungkaliBA(matriks_B,matriks_A)
        
#rumus determinan 2x2
def determinan_2x2(matrix):
    return matrix[0][0]*matrix[1][1] - matrix[1][0]*matrix[0][1]

#rumus determinan 3x3
def determinan_3x3(matrix):
    a, b, c = matrix[0]
    d, e, f = matrix[1]
    g, h, i = matrix[2]
    return a*e*i + b*f*g + c*d*h - c*e*g - b*d*i - a*f*h
def determinan():
    pilihan= input('Pilihan: ')
    if pilihan in ['a', 'A']:
        determinan2x2()
    elif pilihan in ['b', 'B']:
        determinan3x3()
    else:
        print('Pilihan salah.')
        
def determinan2x2():
    print()
    print('Pilih matriks:')
    print('a. Matriks A')
    print('b. Matriks B')
    pilihan= input('Pilihan: ')
    if pilihan in ['a', 'A']:
        determinan2A()
    elif pilihan in ['b', 'B']:
        determinan2B()
    else:
        print('Pilihan salah')
        
def determinan2A():
    if len(matriks_A) != 2:
        print('Error: Dimensi matriks salah.')
        return
    determinanA=determinan_2x2(matriks_A)
    print('Determinan matriks A :',determinanA)
    print()
    
def determinan2B():
    if len(matriks_B) != 2:
        print('Error: Dimensi matriks salah.')
        return
    determinanB=determinan_2x2(matriks_B)
    print('Determinan matriks B :',determinanB)
    print()
def determinan3x3():
    print()
    print('Pilih matriks:')
    print('a. Matriks A')
    print('b. Matriks B')
    pilihan= input('Pilihan: ')
    if pilihan in ['a', 'A']:
        determinan3A()
    elif pilihan in ['b', 'B']:
        determinan3B()
    else:
        print('Pilihan salah')
def determinan3A():
    if len(matriks_A) != 3:
        print('Error: Dimensi matriks salah.')
        return
    determinanA= determinan_3x3(matriks_A)
    print('Determinan matriks A :',determinanA)
    print()
    
def determinan3B():
    if len(matriks_B) != 3:
        print('Error: Dimensi matriks salah.')
        return
    determinanB= determinan_3x3(matriks_B)
    print('Determinan matriks B :',determinanB)
    print()
    
def hitunglagi():
    pilih = input('Kembali ke menu? (y/t)')
    print()
    if pilih in ['y', 'Y', 'yes', 'Yes', 'YES', 'ya', 'YA','Ya']:
        menu()
    else:
        exit("Anda Keluar")
```
